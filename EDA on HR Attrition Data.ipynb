### Bivariate Analysis

```python
ax = sns.countplot(x="Gender",data=df,hue='Attrition')
plt.title("Count of Departure Employees By Gender")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

***From the above graph we can see that Males are more likely to leave then Females***

```python
ax = sns.countplot(x="Department",data=df,hue="Attrition")
plt.title("Department Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

***From the above graphs we can see that Sales and Human Resource department have the most attrition rate***

```python
ax = sns.countplot(x="EnvironmentSatisfaction",data=df,hue='Attrition')
plt.title("Environment-Satisfaction Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

***Environment could be a reason for Employees attrition Low Satisfied Employees are leaving more***

```python
ax = sns.countplot(x="OverTime",data=df,hue='Attrition')
plt.title("Over Time Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

***The attrition rate for employees working overtime stands at 30.53%, which is notably higher than the overall employee attrition rate, highlighting overtime as a potential risk factor***

```python
sns.boxplot(x="Attrition",y="MonthlyIncome",data=df)
plt.title("Monthly Income Vs Attrition")
plt.show()
```

***Most attrition occurs among those earning less than 6,000 units per month. This trend indicates that employees with lower compensation are more likely to leave***

```python
sns.boxplot(x="Attrition",y="Age",data=df)
plt.title("Age Vs Attrition")
plt.show()
```

***The majority of employees who left the company fall within the age range of 28 to 39 years, with a median age of around 32. This suggests that younger to mid-career professionals are more likely to leave***

```python
ax = sns.countplot(x="JobSatisfaction",data=df,hue='Attrition')
plt.title("Job Satisfaction Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

***Employees with Low and Medium job satisfaction score are more likely to leave***

```python
plt.figure(figsize=(22,5))
ax = sns.countplot(x="JobRole",data=df,hue='Attrition')
plt.title("Job Satisfaction Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

***From the above graph we can see that in Sales Representative Job Role the employee attrition rate is higher then other Job Roles***

```python
ax = sns.countplot(x="WorkLifeBalance",data=df,hue='Attrition')
plt.title("Work Life Balance Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

***The attrition rate for employees with Bad work-life-balance stands at 31.25%, which is notably higher than the overall employee attrition rate, highlighting work-life-balance as a potential risk factor***

```python
ax = sns.countplot(x="PerformanceRating",data=df,hue='Attrition')
plt.title("Performance Rating Vs Attrition")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

***These individuals may be leaving due to factors such as lack of career advancement, insufficient recognition, or better external opportunities. This trend emphasizes the need for retention strategies focused on rewarding and engaging top talent, not just measuring their performance***

```python
sns.boxplot(x="Attrition",y="YearsSinceLastPromotion",data=df)
plt.show()
```

***From the above graph we can see that the majority of employees who left the organization had been promoted within the last 0 to 2 years, with a median of around 1 year since their last promotion***

### Correlation Analysis

```python
df["Attrition"] = df["Attrition"].replace('Yes',1)
df["Attrition"] = df["Attrition"].replace('No',0)
```

```python
plt.figure(figsize=(13, 6))
sns.heatmap(df[["Attrition","MonthlyIncome","PerformanceRating","YearsSinceLastPromotion","WorkLifeBalance","YearsAtCompany",
                  'EnvironmentSatisfaction','JobSatisfaction','JobLevel','Education',"TotalWorkingYears",
               "YearsInCurrentRole","YearsWithCurrManager"]].corr(),annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

***From the upper Heatmap we can see that Attrition have only positive correlation with Performance Ratings that means Employees with higher performance ratings are slightly more likely to leave***

### Conclusion

* Departments like "Sales" and "Human Resources" have high attrition rates.
* Attrition rate of Male employees are higher then Female employees.
* Employees with Bad Environment and Job Satisfaction were more likely to leave.
* Job Roles like "Sales Representative" had significantly higher attrition rates.
* Employees with lower monthly income and fewer years since last promotion also showed higher turnover trends.
* Work-Life Balance and overtime hours were also found to influence attrition directly.
